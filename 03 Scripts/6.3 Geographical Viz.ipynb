## Plotting Choropleth

```python
#Creating df with only desired columns
data_to_plot = df[['State','Average Birth Weight (lbs)']]
data_to_plot.head()
```

```python
# Setup a folium map to show average baby weight
map = folium.Map(location = [41,-99], zoom_start = 4)

# Choropleth maps bind Pandas Data Frames and json geometries.This allows us to quickly visualize data combinations
folium.Choropleth(
    geo_data = country_geo, 
    data = data_to_plot,
    columns = ['State', 'Average Birth Weight (lbs)'],
    key_on = 'feature.properties.name', 
    fill_color = 'PuRd', fill_opacity=0.6, line_opacity=0.1,
    legend_name = "Birth Weight").add_to(map)
folium.LayerControl().add_to(map)

map
```

These results are indicating which states on average have the heaviest and lightest babies born. This indicates that Oregon, Idaho, Montana, South Dakota, Nebraska, and New Hampshire have teh heaviest babies born. Where as Arkansas, Mississippi and West Virgina have the lightest babies born. 

```python
# Setup a folium map to show average mother's age
map = folium.Map(location = [41,-99], zoom_start = 4)

# Choropleth maps bind Pandas Data Frames and json geometries.This allows us to quickly visualize data combinations
folium.Choropleth(
    geo_data = country_geo, 
    data = df,
    columns = ['State', 'Average Age of Mother (years)'],
    key_on = 'feature.properties.name', 
    fill_color = 'PuRd', fill_opacity=0.6, line_opacity=0.1,
    legend_name = "Average Age").add_to(map)
folium.LayerControl().add_to(map)

map
```

Interestingly, this shows the average age of mothers across the us. Montana, Nebraska and Vermont haveing the oldest average age of the mother.
